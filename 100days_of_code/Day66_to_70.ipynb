```python
#User Interface Design

import tkinter as tk

window = tk.Tk()
window.title("Hello World")
window.geometry("400x200")

counter = 0
total = 0  # Variable to accumulate the sum

def updateLabel():
    global counter, total
    try:
        number = int(entry.get())  # Get number from Entry
        total += number
        counter += 1
        hello["text"] = f"Clicked {counter} times, Total: {total}"
    except ValueError:
        hello["text"] = "Please enter a valid number"

hello = tk.Label(window, text="Hello World!")
hello.grid(row=0, column=0, columnspan=2, pady=10)

entry = tk.Entry(window, width=25)
entry.grid(row=1, column=0, columnspan=2, pady=5)

button1 = tk.Button(window, text="Click Me!", command=updateLabel)
button1.grid(row=2, column=0, padx=5, pady=5)

button2 = tk.Button(window, text="Another Button!", command=updateLabel)
button2.grid(row=2, column=1, padx=5, pady=5)

button3 = tk.Button(window, text="Last One!", command=updateLabel)
button3.grid(row=3, column=0, columnspan=2, pady=5)
```

```python
import tkinter as tk

window = tk.Tk()
window.title("Calculator")
window.geometry("400x200")

answer = "" 
lastNumber = 0
operator = None

def typeAnswer(value):
    global answer
    answer = f'{answer}{value}'   
    hello["text"] = answer        

def calcAnswer(thisOp):
    global answer, lastNumber, operator 
    if thisOp == "+":
        lastNumber = int(answer)  
        answer = ""               
        operator = "+"           
        hello["text"] = 0         

def calc():
    global answer, lastNumber, operator 
    if operator == "+":
        total = lastNumber + int(answer)  
        hello["text"] = total             

hello = tk.Label(text=answer)
hello.grid(row=0, column=1)

one = tk.Button(text="1", command=lambda: typeAnswer(1))
one.grid(row=1, column=0)

two = tk.Button(text="2", command=lambda: typeAnswer(2))
two.grid(row=1, column=1)

three = tk.Button(text="3", command=lambda: typeAnswer(3))
three.grid(row=1, column=2)

four = tk.Button(text="4", command=lambda: typeAnswer(4))
four.grid(row=2, column=0)

five = tk.Button(text="5", command=lambda: typeAnswer(5))
five.grid(row=2, column=1)

six = tk.Button(text="6", command=lambda: typeAnswer(6))  
six.grid(row=2, column=2)

seven = tk.Button(text="7", command=lambda: typeAnswer(7))
seven.grid(row=3, column=0)

eight = tk.Button(text="8", command=lambda: typeAnswer(8))
eight.grid(row=3, column=1)

nine = tk.Button(text="9", command=lambda: typeAnswer(9))
nine.grid(row=3, column=2)

zero = tk.Button(text="0", command=lambda: typeAnswer(0))
zero.grid(row=4, column=1)

add = tk.Button(text="+", command=lambda: calcAnswer("+")) 
add.grid(row=1, column=4)

equals = tk.Button(text="=", command=calc)
equals.grid(row=4, column=4)

tk.mainloop()
```
